# Transformando um Gráfico em Figura

- aumenta o potêncial dos comandos

```python

# 8 - polegadas
# 4 - polegadas
# # 1 polegada = 2.54 cm
# largura 8 = 20.32 cm
# altura  4 = 10.16 cm
# DPI (dots per inch, ou pontos por polegada).
# Padrão no matplotib - DPI = 100
# Largura : 8 polegadas x 100 DPI = 800 pixels.
# Altura  : 4 polegadas x 100 DPI = 400 pixels.
plt.figure(figsize=(8,4))
plt.plot(df_brasil['ano'],df_brasil['imigrantes'])
plt.title("Imigração do Brasil para o Canadá")
plt.xlabel('Ano',fontsize=9)
plt.ylabel('Número de imigrantes',fontsize=9)
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2013'])
plt.show()
```

# Comparativo dos maiores paises da America Latina (imigração Canadá)

```python
# Criar uma figura
fig,axs = plt.subplots(2,2,figsize=(10,6))

# Brasil
axs[0,0].plot(df.loc['Brasil',anos])
axs[0,0].set_title('Brasil')

# Colombia
axs[0,1].plot(df.loc['Colômbia',anos])
axs[0,1].set_title('Colômbia')

# Argentina
axs[1,0].plot(df.loc['Argentina',anos])
axs[1,0].set_title('Argentina')

# Peru
axs[1,1].plot(df.loc['Peru',anos])
axs[1,1].set_title('Peru')

fig.subplots_adjust(hspace=0.5,wspace=0.6)

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))

for ax in axs.flat:
  ax.set_xlabel('Ano',fontsize=10)
  ax.set_ylabel('Números de Imigrantes',fontsize=10)

# Usando ravel() - achatar a matriz evitando o uso complexo do for
# Usando set_ylim () - definir o limite os valores.
ymin = 0
ymax = 7000
for ax in axs.ravel():
    ax.set_ylim(ymin,ymax)
# suptitle
fig.suptitle("Imigração na América Latina",fontsize=16)
# Customizando o Matplolib (cores, fontes, bordas)
```

# Mostrar os valore (anotações) em cada barra.

# Algoritmo

- Adicionar uma estrutura de repetição-iteração (for)
- Objetivo adicionar os valores máximos de cada barra.
- A primeira linha será a função enumerate() e vai iterar sobre cada item da coluna do valor total.
- Atribuiremos a posição de cada item a uma variável chamada i
- E usaremos a variável v para estar relacionada a cada variável.
  - for i,v in enumerate(america_sul_sorted['Total'])
- Use a função text( ) para adicionar textos as figuras
- ax.text(v+20,i,str(v),color='black',fontsize=10,ha='left',va='center')
- v+20 : indica a posição onde o texto será adicionado, onde o v corresponde a cada uma das barras
- O i é a posição y do nosso texto.
- Por fim, usamos o str( para indicar que vamos converter o camanho de cada barra em uma string, que será o nosso texto)
- Depois disso, passamos a cor e informações do texto


```python
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(12, 7))

cor_brasil = '#ffb6c1'  
cor_padrao = '#D3D3D3'  
cores = [cor_brasil if pais == 'Brasil' else cor_padrao for pais in america_sul_sorted.index]

bars = ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores, height=0.6)

for bar in bars:
    width = bar.get_width()  # Obtém a largura da barra
    # Adiciona o texto ao lado da barra
    ax.text(
        width * 1.02,  # Posição horizontal ajustada
        bar.get_y() + bar.get_height() / 2,  # Posição vertical centralizada
        f'{width:.0f}',  # Texto a ser exibido
        va='center',  # Alinhamento vertical
        ha='left',  # Alinhamento horizontal
        fontsize=10,  # Tamanho da fonte
        color='black'  # Cor do texto
    )


ax.set_title('Imigração da América do Sul para o Canadá\n1980 a 2013', fontsize=18, fontweight='bold')
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')

ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

ax.xaxis.grid(False)


plt.tight_layout()

plt.show()
```
