# Temas do Seaborn

```python
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
```

```python
ax = sns.barplot(data=top_10,y=top_10.index,x='Total',orient='h')
ax.set(
     title='Os dez países com mais imigrantes para o Canadá',
     xlabel='Total de Imigrantes',
     ylabel=''


)
plt.show()
```

```python
def gerar_grafico_paleta(pallete):
    fig,ax = plt.subplots(figsize=(8,4))
    ax = sns.barplot(data=top_10,y=top_10.index,x='Total',
                     orient='h',
                     palette=pallete)
    ax.set_title('Os dez países com mais imigrantes para o Canadá',
                 loc='left',
                 fontsize=16)
    ax.set_xlabel('Total de Imigrantes',fontsize=14)
    ax.set_ylabel('')
    plt.show()
```

# Gráficos Interativos com Plotly

```python
import plotly.express as px
```

```python
anos = list(map(str,range(1980,2014)))
```

```python
brasil = df.loc['Brasil',anos]
```

```python
brasil_dict = {'ano':brasil.index.tolist(),'imigrantes':brasil.values.tolist()}
```

```python
brasil_dict
```

```python
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil.head()
```

```python
# @title imigrantes

from matplotlib import pyplot as plt
dados_brasil['imigrantes'].plot(kind='line', figsize=(8, 4), title='imigrantes')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
fig = px.line(dados_brasil,
              x='ano',
              y='imigrantes',
              title='Imigrantes do Brasíl para o Canada \n de 1980 a 2013')
# centralizar o título geral
fig.update_layout(title_x=0.5)
fig.update_layout(width=1000,
                  height=500,
                  xaxis={'tickangle':-45})
fig.update_traces(line_color='#9751FC',line_width=6)
#fig.update_xaxes(tickangle=-45)
fig.show()
```

```python
fig.write_html('imigrantes_brasil_tarde.html')
```

# Desafio : Redirecionar os xaxis - tickangle

```python
# Artur

fig = px.line(dados_brasil, x="ano", y="imigrantes", title="Imigrantes do Brasil")
fig.update_layout(width=1000, height=500, xaxis_tickangle=-45)
fig.show()
```

```python
# Claudio
fig = px.line(dados_brasil,x='ano',
              y='imigrantes',
              title='Imigrantes do Brasil para o Canada \n de 1980 a 2013')
# Isso aqui gera o gráfico, utilizando o line e passando o Dataframe dados_brasil e a coluna com seus devidos eixos.
fig.update_layout(width=1050,
                  height=600,
                  font_family='Arial',
                  font_color='grey',
                  title_font_color='black',
                  title_font_size=22,
                  xaxis_title='Ano',
                  yaxis_title='Número de Imigrantes')
fig.update_traces(line_color='#2270E3',line_width=3)
fig.update_xaxes(tickangle=0)
fig.update_xaxes(nticks=14)
fig.show()
```

```python
# Salvando Gráficos

fig.write_html('imigrantes_brasil.html')
```

```python
from flask import Flask,render_template
app = Flask(__name__)
@app.route('/graficolinha')
def mostrar_graf_linha():
    return render_template('imigrantes_brasil_tarde.html')

app.run()
```
